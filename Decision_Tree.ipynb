##Decision Tree Classification

```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
#independent  variables
inputs = df.drop('salary_more_then_100k' , axis = 'columns')
```

```python
#Dependent  variables
target = df['salary_more_then_100k']
```

```python
target
```

Now labelled encoding is required in order to use decision tree 

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python

# n represents numbered value or encoded value

# ENCODING IS AS FOLLOWS USING FIT TRANSFORM

inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'], axis = 'columns')
inputs_n
```

```python
target
```

Now using decison tree model from sklearn

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(inputs_n,target)
```

We could also use train split over here but this is just to easy so okay

```python
model.score(inputs_n,target)
```

Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?

```python
model.predict([[2,1,0]])
```

Is salary of Google, Computer Engineer, Masters degree > 100 k ?

```python
model.predict([[2,1,1]])
```

In this scene people who paid high fare had better security than low fare 
so here we have to decide if the person survived or not on the basis of fare and other independent variables


```python
import pandas as pd 
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

Dropping the unnecessary tables

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis = 'columns',inplace = True)
```

```python
df.head()
```

```python
#independent and dependent variables

inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
#simple labelled encoding using python map function

inputs.Sex = inputs.Sex.map({'male':1 , 'female':2})
```

```python
inputs.Age[:10]
```

```python
#filling nan values

inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train ,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

Importing decision tree model

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test
```

```python
model.predict([[3,1,24.000000,7.1417]])
```
